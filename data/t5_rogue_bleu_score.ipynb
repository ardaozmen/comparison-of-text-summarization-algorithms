```python
!pip install rouge

from nltk.translate.bleu_score import sentence_bleu
from rouge import Rouge

rouge = Rouge()
```

```python
for s in range(0,len(raw)):
  print(raw['Actual Text'][s])
  print(raw['Generated Text'][s])
  print('BLEU score -> {}'.format(sentence_bleu([raw['Actual Text'][s].split(' ')], raw['Generated Text'][s].split(' '))))
  print(rouge.get_scores(raw['Generated Text'][s], raw['Actual Text'][s])) 
```
